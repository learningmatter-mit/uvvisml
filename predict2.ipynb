```python
import chemprop
import pandas as pd
```

```python
#Train abs_all with gpu

arguments = [
    '--data_path', 'uvvisml/data/splits/abs_all/deep4chem/group_by_smiles/smiles_target_train.csv',
    '--separate_val_path', f'uvvisml/data/splits/abs_all/deep4chem/group_by_smiles/smiles_target_val.csv',
    '--separate_test_path', f'uvvisml/data/splits/abs_all/deep4chem/group_by_smiles/smiles_target_test.csv',
    '--dataset_type', 'regression',
    '--save_dir', 'uvvisml/models/abs_all_checkpoints/group_by_smiles',
    '--epochs', '30', #100-200
    '--number_of_molecules', '2',
    '--smiles_columns', 'smiles', 'solvent',
    '--gpu', 0
]

args = chemprop.args.TrainArgs().parse_args(arguments)
mean_score, std_score = chemprop.train.cross_validate(args=args, train_func=chemprop.train.run_training)
```

```python
#Test abs all

arguments = [
    '--test_path', f'uvvisml/data/splits/abs_all/deep4chem/group_by_smiles/smiles_target_test.csv',
    '--preds_path', f'uvvisml/models/abs_all_checkpoints/group_by_smiles/abs_all_preds.csv',
    '--checkpoint_dir', f'uvvisml/models/abs_all_checkpoints/group_by_smiles',
    '--number_of_molecules', '2',
    '--smiles_columns', 'smiles', 'solvent'
]

predict_args = chemprop.args.PredictArgs().parse_args(arguments)
preds = chemprop.train.make_predictions(args=predict_args)
```

```python
#Train both peaks
arguments = [
    '--test_path', 'uvvisml/data/splits/both_peaks/deep4chem/group_by_smiles/smiles_target_test.csv',
    '--preds_path', 'uvvisml/models/both_peaks_checkpoints/group_by_smiles/both_peaks_preds.csv',
    '--checkpoint_dir', 'uvvisml/models/both_peaks_checkpoints'
]

args = chemprop.args.PredictArgs().parse_args(arguments)
preds = chemprop.train.make_predictions(args=args)
```

```python
df_both_peaks = pd.read_csv('uvvisml/data/splits/both_peaks/deep4chem/group_by_smiles/smiles_target_test.csv')
df_both_peaks['abs_peakwavs_max_preds'] = [x[0] for x in preds]
df_both_peaks['emi_peakwavs_max_preds'] = [x[1] for x in preds]
df_both_peaks
```

```python
import matplotlib.pyplot as plt

true_value = df['peakwavs_max']
predicted_value = df['preds']

plt.figure(figsize=(5,5))
plt.scatter(true_value, predicted_value)
plt.title('lambda max abs')
plt.xlabel('true values', fontsize=15)
plt.ylabel('predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rms = mean_squared_error(true_value, predicted_value, squared=False)
rms
```

```python
mae = mean_absolute_error(true_value, predicted_value)
mae
```

```python
r2 = r2_score(true_value, predicted_value)
r2
```
